## TPC3: Jogo dos 21 fósforos

Desenvolve em Python o código necessário para o jogo dos 21 fósforos: 

### O jogo

* No início do jogo, há 21 fósforos;
* Cada jogador (computador ou utilizador), pode tirar 1, 2, 3 ou 4 fósforos quando for a sua vez de jogar;
* Os jogadores jogam alternadamente;
* **Quem tirar o último fósforo perde!**

### O programa 

* O jogo deverá ter dois modos: o jogador joga em primeiro lugar e o computador começa a jogar em segundo lugar e, no segundo modo, o computador começa em primeiro; 
* Quando o computador começa a jogar em segundo lugar, deve ganhar sempre o jogo;
* Quando o computador começa a jogar em primeiro lugar, se o utilizador cometer um erro de cálculo, o computador deverá passar para a posição de vencedor e ganhar o jogo.

```python
def jogador(restantes):
    jogada = 0
    while jogada < 1 or jogada > 4 or jogada > restantes:
        jogada = int(input(f"Fósforos restantes: {restantes}. Quantos fósforos quer remover de 1 a 4? "))
        if jogada < 1 or jogada > 4 or jogada > restantes: # caso numero nao esteja no intervalo de 1 a 4 
            print("Jogada inválida! Escolha um número entre 1 e 4 que seja possível.")
    return jogada

# remove sempre 4
def computador(restantes):
    if restantes > 4:
        jogada = 4
    else:
        jogada = restantes - 1  # Tentar sempre deixar 1 fósforo para o jogador
    print(f"Computador remove {jogada} fósforos.")
    return jogada

# Jogador começa
def modo_1():
    fosforos = 21
    print("\nModo 1:\n O jogador começa e o computador joga depois.")

    while fosforos > 0:
        # Jogada do jogador
        jogada_jogador = jogador(fosforos) #chama sempre com o numero de fosforos resantes 
        fosforos = fosforos - jogada_jogador
        if fosforos == 0:
            print("Perdeste, o computador ganhou :")
            return
        
        # Jogada do computador
        jogada_computador = computador(fosforos)
        fosforos = fosforos - jogada_computador
        if fosforos == 0:
            print("O computador perdeu! Você venceu.")
            return

# Computador começa
def modo_2():
    fosforos = 21
    print("\nModo 2:\n O computador começa a jogar.")

    while fosforos > 0:
        # Jogada do computador
        jogada_computador = computador(fosforos)
        fosforos = fosforos - jogada_computador
        if fosforos == 0:
            print("O computador perdeu! You won.")
            return

        # Jogada do jogador
        jogada_jogador = jogador(fosforos)
        fosforos = fosforos - jogada_jogador
        if fosforos == 0:
            print("Você perdeu! O computador venceu.")
            return

# Menu principal para escolher o modo de jogo
def menu():
    print("Bem-vindo ao jogo dos 21 fósforos!")
    print("Escolha o modo de jogo:")
    print("1. Jogador começa.")
    print("2. Computador começa.")

    escolha = ""
    while escolha not in ["1", "2"]:
        escolha = input("Digite 1 ou 2: ")
    
    if escolha == "1":
        modo_1()
    else:
        modo_2()

if __name__ == "__main__":
    menu()
```
